```python
ANS-1


df.plot(): This method is used to create various types of plots like line plots, bar plots, scatter plots, etc., directly from a DataFrame.

import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Year': [2015, 2016, 2017, 2018, 2019],
        'Sales': [1000, 1200, 1400, 1600, 1800]}

# Create DataFrame
df = pd.DataFrame(data)

# Create a line plot
df.plot(x='Year', y='Sales', kind='line', marker='o')
plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()



ANS-2


import pandas as pd

def reindex_dataframe(df):
    # Function to generate new index
    def generate_new_index(index):
        return (index * 2) + 1

    # Reset the index and apply the custom function
    df_reset = df.reset_index(drop=True)
    df_reset.index = df_reset.index.map(generate_new_index)

    return df_reset

# Sample DataFrame
data = {'A': [10, 20, 30, 40],
        'B': [50, 60, 70, 80],
        'C': [90, 100, 110, 120]}

df = pd.DataFrame(data)

# Re-index the DataFrame with the new index
new_df = reindex_dataframe(df)

print(new_df)



ANS-3




import pandas as pd

def calculate_sum_first_three(df):
    sum_first_three = df['Values'].head(3).sum()
    print("Sum of the first three values:", sum_first_three)

# Sample DataFrame
data = {'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Calculate and print the sum of the first three values in the 'Values' column
calculate_sum_first_three(df)




ANS-4



import pandas as pd

def count_words(df):
    # Custom function to count words in each row of 'Text' column
    def word_count(text):
        return len(text.split())

    # Apply the custom function to create a new column 'Word_Count'
    df['Word_Count'] = df['Text'].apply(word_count)

    return df

# Sample DataFrame
data = {'Text': ["Hello, how are you?", "Python programming is fun", "Data analysis using Pandas"]}
df = pd.DataFrame(data)

# Count words and create a new 'Word_Count' column
df_with_word_count = count_words(df)

print(df_with_word_count)




ANS-5


DataFrame.size():
DataFrame.size() is a method that returns the total number of elements in the DataFrame.
It calculates the size by multiplying the number of rows (length of the DataFrame) by the number of columns.
The size includes all the elements in the DataFrame, regardless of whether they contain NaN or valid data.
It returns a single integer value representing the total number of elements in the DataFrame.
Example:
    
    
    import pandas as pd

data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)

print("DataFrame size:", df.size)  # Output: 6 (3 rows * 2 columns = 6)



ANS-6



Here's the syntax of the pd.read_excel() function:


import pandas as pd

df = pd.read_excel('file_path.xlsx', sheet_name='sheet_name')



Parameters:

file_path: The file path or URL of the Excel file you want to read.
sheet_name: (Optional) The name or index of the sheet within the Excel file to read. If not specified, it will read the first sheet by default.
Example:
    
    
    
       Name  Age  Score
0   Amy   25     90
1   Ben   30     85
2  Jack   22     95



You can read this Excel file using pandas as follows:
    
    import pandas as pd

# Read the Excel file
df = pd.read_excel('data.xlsx')

# Display the DataFrame
print(df)


   Name  Age  Score
0   Amy   25     90
1   Ben   30     85
2  Jack   22     95



ANS-7


import pandas as pd

def extract_username(df):
    # Extract username from email using str.split() and str.get() methods
    df['Username'] = df['Email'].str.split('@').str.get(0)

    return df

# Sample DataFrame
data = {'Email': ['john.doe@example.com', 'jane.smith@example.com', 'bob.johnson@example.com']}
df = pd.DataFrame(data)

# Extract the username from email and create a new 'Username' column
df_with_username = extract_username(df)

print(df_with_username)



                    Email   Username
0    john.doe@example.com   john.doe
1  jane.smith@example.com  jane.smith
2  bob.johnson@example.com  bob.johnson



The function extract_username() takes the DataFrame df, uses the str.split('@') method to split the email addresses by the '@' symbol, and then uses the str.get(0) method to get the first part of the split, which is the username. It creates a new column 'Username' in the DataFrame to store the extracted usernames.




ANS-8



import pandas as pd

def select_rows(df):
    # Use boolean indexing to select rows based on conditions
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]

    return selected_rows

# Sample DataFrame
data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}
df = pd.DataFrame(data)

# Call the function to select rows based on conditions
selected_df = select_rows(df)

print(selected_df)


   A  B  C
1  8  2  7
4  9  1  2



ANS-9



import pandas as pd

def calculate_statistics(df):
    # Calculate mean, median, and standard deviation of the 'Values' column
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_value = df['Values'].std()

    return mean_value, median_value, std_value

# Sample DataFrame
data = {'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Calculate statistics
mean_value, median_value, std_value = calculate_statistics(df)

print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_value)



Mean: 30.0
Median: 30.0
Standard Deviation: 15.811388300841896





ANS-10




import pandas as pd

def calculate_moving_average(df):
    # Convert 'Date' column to datetime type (if not already)
    df['Date'] = pd.to_datetime(df['Date'])

    # Sort DataFrame by 'Date' in ascending order (if not already sorted)
    df = df.sort_values(by='Date')

    # Calculate the moving average using a window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df

# Sample DataFrame
data = {'Date': ['2023-07-01', '2023-07-02', '2023-07-03', '2023-07-04', '2023-07-05', '2023-07-06', '2023-07-07', '2023-07-08'],
        'Sales': [100, 120, 130, 110, 140, 150, 160, 170]}
df = pd.DataFrame(data)

# Call the function to calculate the moving average
df_with_moving_average = calculate_moving_average(df)

print(df_with_moving_average)


OUTPUT

        Date  Sales  MovingAverage
0 2023-07-01    100     100.000000
1 2023-07-02    120     110.000000
2 2023-07-03    130     116.666667
3 2023-07-04    110     115.000000
4 2023-07-05    140     120.000000
5 2023-07-06    150     125.000000
6 2023-07-07    160     131.428571
7 2023-07-08    170     138.750000



ANS-11



import pandas as pd

def add_weekday_column(df):
    # Convert 'Date' column to datetime type (if not already)
    df['Date'] = pd.to_datetime(df['Date'])

    # Extract weekday names and create the 'Weekday' column
    df['Weekday'] = df['Date'].dt.strftime('%A')

    return df

# Sample DataFrame
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
df = pd.DataFrame(data)

# Call the function to add the 'Weekday' column
df_with_weekday = add_weekday_column(df)

print(df_with_weekday)




        Date    Weekday
0 2023-01-01     Sunday
1 2023-01-02     Monday
2 2023-01-03    Tuesday
3 2023-01-04  Wednesday
4 2023-01-05   Thursday




ANS-12




import pandas as pd

def select_rows_between_dates(df):
    # Convert 'Date' column to datetime type (if not already)
    df['Date'] = pd.to_datetime(df['Date'])

    # Create start and end date for filtering
    start_date = pd.to_datetime('2023-01-01')
    end_date = pd.to_datetime('2023-01-31')

    # Use boolean indexing to select rows based on conditions
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

    return selected_rows

# Sample DataFrame
data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-20', '2023-02-05']}
df = pd.DataFrame(data)

# Call the function to select rows between '2023-01-01' and '2023-01-31'
selected_df = select_rows_between_dates(df)

print(selected_df)




OUTPUT


        Date
0 2023-01-01
1 2023-01-15
2 2023-01-20



ANS-13



To use the basic functions of pandas, the first and foremost necessary library that needs to be imported is pandas itself. The pandas library provides the fundamental data structures and functions for data manipulation and analysis, including the DataFrame and Series objects.

You can import the pandas library with the following import statement:
    
    
    
    import pandas as pd




```
